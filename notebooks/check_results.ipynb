```python
%matplotlib inline
from astropy.io import ascii
from gwpy.timeseries import TimeSeries
import numpy as np
```

```python
tab = ascii.read('../fit_table.dat')
st = np.asarray(tab['starttime'])
```

```python
amplitudes0 = TimeSeries(tab['amp0'], x0=st[0], dx=st[2] - st[1])
peaks0 = TimeSeries(tab['peaks0'], x0=st[0], dx=st[2] - st[1])
```

```python
plot = amplitudes0.asd().plot(figsize=(12,6))
ax = plot.gca()
ax.set_xscale('log')
ax.plot([1./86400, 1./86400], [1e-4, 1e-1], c='C1')
ax.plot([2./86400, 2./86400], [1e-4, 1e-1], c='C1')
ax.set_ylim(1e-4, 1e-1)
plot.show()
```
